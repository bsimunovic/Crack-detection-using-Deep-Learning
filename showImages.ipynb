```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import jinja2
import cv2
```

```python
data_dir_train_cracked = 'Dataset/cracked.csv'
data_dir_train_uncracked = 'Dataset/uncracked.csv'

data_dir_train_cracked_thresholding = 'Dataset/Tcracked.csv'
data_dir_train_uncracked_thresholding = 'Dataset/Tuncracked.csv'

data_dir_train_cracked_thresholdingCanny = 'Dataset/TCcracked.csv'
data_dir_train_uncracked_thresholdingCanny = 'Dataset/TCuncracked.csv'

data_dir_train_cracked_ImageSegmentation = 'Dataset/TCKcracked.csv'
data_dir_train_uncracked_ImageSegmentation = 'Dataset/TCKuncracked.csv'

Cracked_Img = pd.read_csv(data_dir_train_cracked)
Uncracked_Img = pd.read_csv(data_dir_train_uncracked)

Cracked_Img_Thresholding = pd.read_csv(data_dir_train_cracked_thresholding)
Uncracked_Img_Thresholding = pd.read_csv(data_dir_train_uncracked_thresholding)

Cracked_Img_ThresholdingCanny = pd.read_csv(data_dir_train_cracked_thresholdingCanny)
Uncracked_Img_ThresholdingCanny = pd.read_csv(data_dir_train_uncracked_thresholdingCanny)

Cracked_Img_ImageSegmentation = pd.read_csv(data_dir_train_cracked_ImageSegmentation)
Uncracked_Img_ImageSegmentation = pd.read_csv(data_dir_train_uncracked_ImageSegmentation)
```

```python
figure,axis = plt.subplots(1,4,figsize=(12,12))

for indexing, plot in enumerate(axis.flat):
    
    Reading_Img = cv2.imread(Cracked_Img["PATH"][indexing])
    Reading_Img = cv2.cvtColor(Reading_Img,cv2.COLOR_BGR2RGB)
    
    plot.set_title("Cracked")
    plot.set_xlabel(Reading_Img.shape)
    plot.set_ylabel(Reading_Img.size)
    plot.imshow(Reading_Img)
    
plt.tight_layout()   
plt.show()
```
